```python
from sklearn.model_selection import cross_validate, KFold
from sklearn.metrics import make_scorer, accuracy_score, precision_score, recall_score, f1_score
```

```python
k_fold = KFold(n_splits=8, random_state=1, shuffle=True)

score = {'f1_score' : make_scorer(f1_score)}

k_fold_result = cross_validate(rf, x5, y5, cv=k_fold, scoring = score)

k_fold_result.keys()

f1_score = k_fold_result["test_f1_score"].mean()
print("f1_score : {0: .4f}".format(f1_score))
```
